# handling missing values

```python
data = data.dropna(subset=['Total Electric Vehicle', 'Total Non-Electric Vehicle', 'Total'])
```

# data evaluation

```python
data['Total Electric Vehicle'] = data['Total Electric Vehicle'].str.replace(',', '').astype(int)
data['Total Non-Electric Vehicle'] = data['Total Non-Electric Vehicle'].str.replace(',', '').astype(int)
data['Total'] = data['Total'].str.replace(',', '').astype(int)

# Calculate the percentage

data['percentage'] = data['Total Electric Vehicle'] / data['Total']
print("\n\n state name with percentage of EV Vehilces")

print(data[["State Name","percentage"]])
data
```

# Plotting segmented data

```python
# Plotting the data on separate axes
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10), sharex=True)

# Plotting Total Electric Vehicle on the first axis
ax1.bar(data['State Name'], data['Total Electric Vehicle'], label='Total Electric Vehicle', color='blue')
ax1.set_ylabel('Number of Electric Vehicles')
ax1.set_yscale('log')
ax1.legend()

# Plotting Total Non-Electric Vehicle on the second axis
ax2.bar(data['State Name'], data['Total Non-Electric Vehicle'], label='Total Non-Electric Vehicle', color='orange')
ax2.set_ylabel('Number of Non-Electric Vehicles')
ax2.set_yscale('log')
ax2.legend()

plt.xticks(rotation=90)
plt.xlabel('State Name')
plt.title('Total Electric and Non-Electric Vehicles for Each State')
plt.tight_layout()
plt.show()
```

States with Higher EV Usage:
Uttar Pradesh has the highest number of electric vehicles (3,37,180), followed by Delhi (1,56,393), Karnataka(1,20,532),Maharashtra (1,16,646) and Bihar(83335). 

The top five states with higher EV usage (in terms of the number of electric vehicles) are: 
Uttar Pradesh, Delhi, Karnataka, Maharashtra, and Bihar.

States with Lower EV Usage:
Arunachal Pradesh has the lowest number of electric vehicles, followed by Sikkim and Mizoram (21), Ladakh(26), and Meghalaya (49). The bottom five states with lower EV usage are Arunachal Pradesh, Mizoram, Sikkim, Ladakh, and Meghalaya.

States with Lower EV Usage but Higher Non-EV Usage:
Mizoram has relatively lower electric vehicle usage compared to some other states but has a higher number of non-electric vehicles. Similarly, Arunachal Pradesh and Meghalaya also have lower EV usage but a relatively higher number of non-electric vehicles.

States with Higher EV Usage and Lower Non-EV Usage:
Delhi has relatively higher electric vehicle usage compared to some states and lower non-electric vehicle usage. Tripura, Uttar Pradesh, and Karnataka also have higher EV usage and lower non-electric vehicle usage. 

States with Balanced EV and Non-EV Usage:
Some states like Assam, Gujarat, Haryana, Kerala, Odisha, Punjab, and West Bengal have a significant number of electric vehicles and a considerable number of non-electric vehicles. 


States with Missing Data:
Some states like Andra Pradesh, Lakshadweep, Madhya Pradesh, and Telangana have missing data for both electric and non-electric vehicles.
